```
from scipy.stats import norm
import matplotlib.pyplot as plt
import numpy as np
```

```
'''
common methods
rvs: Random Variates pdf: Probability Density Function cdf: Cumulative 
Distribution Function sf: Survival Function (1-CDF) ppf: Percent Point Function 
(Inverse of CDF) isf: Inverse Survival Function (Inverse of SF) stats: Return 
mean, variance, (Fisher’s) skew, or (Fisher’s) kurtosis moment: non-central 
moments of the distribution Let’s take a normal RV as an example.
'''
# create a subplot
fig, ax = plt.subplots(1,1)

# mean variance skew and kurtios
mean, var, skew, kurt = norm.stats(moments = 'mvsk')

# display and probability density function
x = np.linspace(norm.ppf(0.01), norm.ppf(0.99), 100)
ax.plot(x, norm.pdf(x), 'r-', lw = 5, alpha = 0.6, label = 'norm pdf' )

# Freeze the distribution and display the frozen pdf:
rv = norm()
ax.plot(x, rv.pdf(x), 'k-', lw = 2, label = 'frozen pdf')
r = rv.rvs(size = 1000)

ax.hist(r, density=True, histtype='stepfilled', alpha = 0.2)
ax.legend(loc = 'best', frameon = True)
plt.show();
```
